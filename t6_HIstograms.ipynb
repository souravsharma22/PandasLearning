```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore","use_inf_as_na")
iris = sns.load_dataset("iris")
print(iris)
```

```python
#Ploting this on histagram
```

```python
ax= iris.plot.hist(column =["sepal_length"],bins=20,grid = True)
ax.set_title("Histagram of the sepal_length of 150 flowers",fontsize = 16)
ax.set_xlabel("sepal_length in cm",fontsize = 12)
ax.set_ylabel("count",fontsize= 12)
plt.show()
```

```python
#Similarly we can make histagram for other column(Properties)
#More the value of bin more detailed data will be 
ax= iris.plot.hist(column =["sepal_length"],bins=50,grid = True)
ax.set_title("Histagram of the sepal_length of 150 flowers",fontsize = 16)
ax.set_xlabel("sepal_length in cm",fontsize = 12)
ax.set_ylabel("count",fontsize= 12)
plt.show()
```

```python
fig , ax = plt.subplots()
colors  = {"setosa":"Red","versicolor":"Blue","virginica":"Yellow"}
for species , data in iris.groupby("species"):
    data["sepal_length"].plot(kind="hist",column="sepal_length",bins= 15,alpha = 0.50,color=colors[species],ax = ax)
ax.set_xlabel("sepal_length in cm",fontsize= 12)
ax.set_ylabel("count",fontsize = 12)
ax.legend(colors.keys(),loc="upper left")
fig.suptitle("")
ax.set_title("Histagram of the sepal_length of 150 flowers",fontsize = 16)
plt.show()
```

```python
sns.histplot(data=iris,x = "sepal_length",y= "sepal_width",bins=25,hue="species")
plt.plot()
```

```python
graph = sns.JointGrid()
sns.histplot(data=iris,x = "sepal_length",y = "sepal_width",bins=20,hue= "species",alpha = 0.60,ax = graph.ax_joint)
sns.histplot(data=iris,x= "sepal_length",bins= 20,hue="species",alpha =0.60,ax= graph.ax_marg_x)
sns.histplot(data=iris,y= "sepal_width",bins= 20,hue="species",alpha =0.60,ax= graph.ax_marg_y)
graph.figure.subplots_adjust(top=0.90)
graph.figure.suptitle(" 4 histagram joint grid plot for 2 variables in the Iris DataSet")
plt.show()
```
