```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore","use_inf_as_na")
iris = sns.load_dataset("iris")
print(iris)
```

```python
ax = iris.plot.scatter(x="sepal_length",y="sepal_width",color= "Blue")
ax.set_xlabel("sepal length in cm",fontsize = 12)
ax.set_ylabel("sepal width in cm",fontsize = 12)
ax.set_title("sepal data fow flowers in cm",fontsize = 16)
plt.show()
```

```python
fig , ax = plt.subplots()
colors={"setosa":"Red","versicolor":"Blue","virginica":"Yellow"}
for species ,data in iris.groupby("species"):
    data.plot(kind="scatter", x = "sepal_length",y = "sepal_width",label =species,color = colors[species],ax = ax)
ax.set_xlabel("sepal length in cm",fontsize = 12)
ax.set_ylabel("sepal width in cm",fontsize = 12)
ax.set_title("sepal data fow flowers in cm with three species",fontsize = 16)
plt.show()
```

```python
fig , ax = plt.subplots()
colors={"setosa":"Red","versicolor":"Blue","virginica":"Yellow"}
for species ,data in iris.groupby("species"):
    data.plot(kind="scatter", x = "sepal_length",y = "sepal_width",label =species,color = colors[species],marker= 'X',ax = ax)
ax.set_xlabel("sepal length in cm",fontsize = 12)
ax.set_ylabel("sepal width in cm",fontsize = 12)
ax.set_title("sepal data fow flowers in cm with three species",fontsize = 16)
plt.show()
```

```python
fig , ax = plt.subplots()
colors={"setosa":"Red","versicolor":"Blue","virginica":"Yellow"}
for species ,data in iris.groupby("species"):
    data.plot(kind="scatter", x = "sepal_length",y = "sepal_width",label =species,color = colors[species],marker= 'X',ax = ax)
for species ,data in iris.groupby("species"):
    data.plot(kind="scatter", x = "petal_length",y = "petal_width",label =species,color = colors[species],marker= '+',ax = ax)
ax.set_xlabel("sepal length in cm",fontsize = 12)
ax.set_ylabel("sepal width in cm",fontsize = 12)
ax.set_title("sepal data fow flowers in cm with three species",fontsize = 16)
plt.show()
```

```python
#Maling Sctterplot using Seaborn
sns.scatterplot(data=iris,x = "sepal_width",y ="sepal_length",hue="species",style="species")
plt.show()
#In this the color for diffrent species sepal_width will be diffrent
# and as style hab also value "species" hence all species plot will have a diffrent ploting point style
```

```python
# we can have diffrent type of palette,palette is used for color control
sns.scatterplot(data=iris,x = "sepal_width",y ="sepal_length",hue="species",style="species",palette="husl")
plt.show()
```

```python
sns.pairplot(data=iris,hue="species")
plt.show()
#this will give plot for every possible combination of the variables
```

```python
iris["sepal"]="sepal"
iris["petal"]="petal"
iris_1 = pd.DataFrame(data=iris[["sepal_length","sepal_width","species","sepal"]])
iris_1.columns = ["length","width","species","measured"]
iris_2 = pd.DataFrame(data=iris[["petal_length","petal_width","species","petal"]])
iris_2.columns = ["length","width","species","measured"]
#print(iris_1)
#print(iris_2)
iris2 = pd.concat([iris_1,iris_2])
#print(iris2)
```

```python
#Now plotting the new iris2 data on scatter style
sns.scatterplot(data=iris2,x= "width",y="length",hue="species",style="measured")
plt.show()
```
