```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

iris = sns.load_dataset("iris")
print(iris)
```

```python
ax = iris.boxplot(column="sepal_length",showmeans = True,meanline = True,grid=False)
plt.suptitle("")
ax.set_title("Box plot for sepal length of 150 flowers",fontsize = 16)
plt.show()
```

```python
ax = iris.boxplot(column="sepal_length", by= "species",showmeans = True,meanline = True,grid=False)
plt.suptitle("")
ax.set_title("Box plot for sepal length of 150 flowers with species",fontsize = 16)
plt.show()
```

```python
sns.boxplot(data=iris,x="species",y = "sepal_length",hue="species")
plt.show()
```

```python
sns.boxplot(data=iris,x="species",y = "sepal_length",hue="species",dodge=False,width=0.3)
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore","use_inf_as_na")
sns.boxenplot(data=iris,x="species",y= "sepal_length",hue="species",dodge=False,width=0.4)
plt.show()
```

```python
sns.violinplot(data=iris,x="species",y = "sepal_length",hue="species",dodge=False,width=0.4)
plt.show()
```

```python
sns.swarmplot(data=iris,x = "species",y ="sepal_length", hue="species")
plt.show()
```

```python
sns.catplot(data=iris,x="species",y = "sepal_length",hue="species",kind="boxen",dodge=False,width=0.4)
sns.swarmplot(data=iris,x = "species",y ="sepal_length", hue="species")
plt.show()
```
