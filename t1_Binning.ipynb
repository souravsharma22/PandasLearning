```python
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data = sns.load_dataset("geyser")
print(data)
```

```python
sns.histplot(data=data,x="waiting",bins=25,hue="kind")
plt.show()
sns.histplot(data=data,x="duration",bins=25,hue="kind")
plt.show()
```

```python
data["du1"]= pd.qcut(data.duration,q=2,labels=["Short","long"])
data["du2"]= pd.qcut(data.duration,q=3,labels=["Short","Medium","long"])
print(data)
```

```python
fig , ((h_dur,du1,du2))= plt.subplots(1,3)
sns.histplot(data=data,x="duration",bins=25,hue="kind",ax=h_dur)
sns.histplot(data=data,x="duration",bins=25,hue="du1",ax=du1)
sns.histplot(data=data,x="duration",bins=25,hue="du2",ax=du2)
plt.show()
```

```python
data["cut1"]= pd.cut(data.duration,bins=2, labels=["Short","Long"])
data["cut2"]= pd.cut(data.duration,bins=3,labels=["Short","Medium","long"])
print(data)
```

```python
fig , ((h_dur,du1,du2),(h_dur2,cut1,cut2))= plt.subplots(2,3)
sns.histplot(data=data,x="duration",bins=25,hue="kind",ax=h_dur)
sns.histplot(data=data,x="duration",bins=25,hue="du1",ax=du1)
sns.histplot(data=data,x="duration",bins=25,hue="du2",ax=du2)
sns.histplot(data=data,x="duration",bins=25,hue="kind",ax=h_dur2)
sns.histplot(data=data,x="duration",bins=25,hue="cut1",ax=cut1)
sns.histplot(data=data,x="duration",bins=25,hue="cut2",ax=cut2)
plt.show()
```

```python
data["scut1"]= pd.cut(data.duration,bins=[1,2.5,100],labels=["Short","Long"])
data["scut2"]= pd.cut(data.duration,bins=[1,2.5,4,100],labels=["Short","Medium","Long"])
print(data)
```

```python
fig , ((h_dur,du1,du2),(h_dur2,cut1,cut2),(h_dur3,s1,s2))= plt.subplots(3,3)
sns.histplot(data=data,x="duration",bins=25,hue="kind",ax=h_dur,legend=None)
sns.histplot(data=data,x="duration",bins=25,hue="du1",ax=du1,legend=None)
sns.histplot(data=data,x="duration",bins=25,hue="du2",ax=du2,legend=None)
sns.histplot(data=data,x="duration",bins=25,hue="kind",ax=h_dur2,legend=None)
sns.histplot(data=data,x="duration",bins=25,hue="cut1",ax=cut1,legend=None)
sns.histplot(data=data,x="duration",bins=25,hue="cut2",ax=cut2,legend=None)
sns.histplot(data=data,x="duration",bins=25,hue="kind",ax=h_dur3,legend=None)
sns.histplot(data=data,x="duration",bins=25,hue="scut1",ax=s1,legend=None)
sns.histplot(data=data,x="duration",bins=25,hue="scut2",ax=s2,legend=None)
plt.show()
```
